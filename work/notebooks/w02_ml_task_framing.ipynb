## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
from IPython.display import Markdown, display

display(Markdown("""
# ML Task Type: Classification

The goal is to identify whether a content item is likely to decline in performance.

Decision improved:
Editors can prioritize which content should be reviewed and refreshed first.

Why classification?
The output is an observed label:
- Declining
- Not Declining

The prediction helps prioritize content refresh decisions rather than manually reviewing thousands of pages.
"""))
```

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

```python
import pandas as pd
from IPython.display import display, Markdown

df = pd.DataFrame({
    "content_id":["C001","C002","C003","C004","C005"],
    "client_id":["A1","A1","B2","B2","C3"],
    "content_type":["Blog","Landing","Blog","Guide","Product"],
    "ctr":[0.82,1.10,0.67,0.94,0.71],
    "avg_position":[12.4,8.7,19.5,10.2,15.1],
    "is_declining_label":[1,0,1,0,1]
})

display(Markdown("""
# Unit of Analysis

Each row represents one content item.

Each column contains measurements describing that content, such as CTR, average search position, content type, and the observed decline label.

The model learns patterns from individual content records rather than from clients as a whole.
"""))

display(df)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
display(Markdown("""
# Why Machine Learning Beats a Fixed Rule

A simple rule such as:

"If CTR is below 1%, refresh the page."

is not reliable.

Content performance depends on many interacting factors including:

- CTR
- Average search position
- Content type
- User engagement
- Historical traffic
- Client differences

Machine Learning can discover complex patterns across these variables that cannot be captured with a single if-statement.

The model supports editors by prioritizing pages for review. Final editorial decisions remain with humans.
"""))
```
